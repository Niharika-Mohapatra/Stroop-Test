```python
#Importing libraries
import pandas as pd 
import numpy as np

import scipy
from scipy.stats import ttest_ind
from scipy.stats import ttest_rel
from scipy.stats import wilcoxon

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Creating dataframe
df = pd.read_csv("stroop_results_final.csv") 
df = df.drop(columns=df.columns[0])
df["is_correct"] = df["is_correct"].astype(int)

#Checking for missing data
assert not df.isnull().values.any(), "Missing values found in the dataset."

print(df.columns)
print(f"Total number of results: {len(df)}")
```

**NOTE: Participant IDs were not originally recorded.**
I assume that each block of 20 trials corresponds to one participant (in the order data was collected). However, this method relies on the data being unshuffled and complete.

```python
#Modifying dataframe 

#Dividing based on id
df["participant_id"] = df.index // 20

#Adding congruent column
df["is_congruent"] = df["word"].str.lower() == df["color"]
df["is_congruent"] = df["is_congruent"].astype(int)

#Filtering invalid values 
df = df[(df["reaction_time"] >= 200) & (df["reaction_time"] <= 3000)]

#Modified dataframe
print(df.head())
print(len(df))
```

**Raw Trial Level Overview:** 

```python
reaction_times = df["reaction_time"]
mean_rt = np.mean(reaction_times)
median_rt = np.median(reaction_times)
std_rt = np.std(reaction_times)

accuracy = df["is_correct"].mean() * 100

print(f"Mean Reaction Time: {mean_rt:.3f} ms")
print(f"Median Reaction Time: {median_rt:.3f} ms")
print(f"Standard Deviation: {std_rt:.3f} ms")
print(f"Percentage Accuracy (Ungrouped): {accuracy:.3f}%")
```

```python
#Histogram of raw data
plt.figure(figsize=(10, 5))
sns.histplot(df["reaction_time"], bins=50, kde=True, color="#3776ab")
plt.axvline(mean_rt, color='red', linestyle='--', label='Mean')
plt.axvline(median_rt, color='green', linestyle='--', label='Median')
plt.xlabel("Reaction Time (ms)")
plt.ylabel("Frequency")
plt.title("Distribution of Reaction Times")
plt.legend()
plt.show() 
```

```python
#Boxplot of raw data
plt.figure(figsize=(4, 6))
sns.boxplot(y=df["reaction_time"], color="#3776ab")
plt.ylabel("Reaction Time (ms)")
plt.title("Boxplot of Reaction Times")
plt.show()
```

**Participant Level Summary:**

```python
#Summary stats per participant
participant_summary = df.groupby("participant_id").agg(
    mean_rt=("reaction_time", "mean"),
    std_rt=("reaction_time", "std"),
    accuracy=("is_correct", "mean")
).reset_index() 
#summary = summary.rename(columns={"is_correct":"accuracy"})
print(participant_summary) 
```

## **Accuracy vs Reaction Time (Speed-Accuracy Tradeoff)

I analyzed whether faster responses were associated with lower accuracy, both across participants and within individual trials. 

### Participant-level:

I examined each participants mean reaction time and mean accuracy to identify broad patterns. 

```python
#Correlation between reaction time and accuracy per participant
corr2 = participant_summary["mean_rt"].corr(participant_summary["accuracy"])
print(f"Participant-level correlation between reaction time and accuracy: {corr2:.3f}")
```

```python
#Scatter plot
sns.scatterplot(data=participant_summary, x='mean_rt', y='accuracy')
plt.title("Accuracy vs. Reaction Time (per participant)")
plt.xlabel("Mean Reaction Time (ms)")
plt.ylabel("Mean Accuracy")
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

**Interpretation:** The participant-level correlation (r = 0.041) was small and not statistically significant, suggesting that participants who responded faster were not meaningfully less accurate in this task.

### Trial level:

I also assessed whether accuracy varied by response at the trial level.

```python
#Correlation between reaction time and accuracy
corr1 = df['reaction_time'].corr(df['is_correct'])
print(f"Trial-level correlation between reaction time and accuracy: {corr1:.3f}")
```

```python
#Boxplot
sns.boxplot(x='is_correct', y='reaction_time', data=df, palette={"0": "#e57373", "1": "#81c784"})
plt.title("Reaction Time by Accuracy")
plt.xticks([0, 1], ["Incorrect", "Correct"])
plt.ylabel("Reaction Time (ms)")
plt.grid(True, linestyle='--', alpha=0.4)
plt.show() 
```

**Interpretation:** On a trial-by-trial basis, a weak positive correlation (r = 0.10) was observed between reaction time and accuracy. This suggests that slightly slower responses tended to be marginally more accurate, though the effect was small and likely not practically significant.

Interestingly, incorrect responses had slightly longer reaction times than correct ones. This runs counter to the typical speedaccuracy tradeoff, where errors often stem from rushed responses. In this case, the longer times may reflect uncertainty or hesitation before committing to an incorrect choice  suggesting confusion rather than impulsivity as the underlying cause of errors.

## Congruent vs Incongruent:

### Participant level:

```python
congruent_means = df[df["is_congruent"] == 1].groupby("participant_id")["reaction_time"].mean()
incongruent_means = df[df["is_congruent"] == 0].groupby("participant_id")["reaction_time"].mean()
print(congruent_means, incongruent_means)
```

```python
#Histograms of congruent vs incongruent stimuli results
plt.figure(figsize=(10, 5))
sns.histplot(df[df["is_congruent"] == 1]["reaction_time"], bins=50, color="blue", label="Congruent", kde=True)
sns.histplot(df[df["is_congruent"] == 0]["reaction_time"], bins=50, color="orange", label="Incongruent", kde=True)
plt.title("Distribution of Reaction Times: Congruent vs Incongruent")
plt.xlabel("Reaction Time (ms)")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend()
plt.show()
```

```python
# Ensuring both groups have the same participants
valid_participants = congruent_means.index.intersection(incongruent_means.index)

congruent_means = congruent_means.loc[valid_participants]
incongruent_means = incongruent_means.loc[valid_participants]

# Paired t-test
t_stat, p_value = ttest_rel(incongruent_means, congruent_means)
print(f"Paired T-Test: t = {t_stat:.3f}, p = {p_value:.3f}") 
```

```python
#Boxplots of congruent vs incongruent stimuli results
grouped = df.groupby(['participant_id', 'is_congruent'])['reaction_time'].mean().reset_index()

sns.boxplot(x='is_congruent', y='reaction_time', data=grouped, palette="Set2")
plt.title("Mean Reaction Time per Participant by Condition")
plt.ylabel("Mean Reaction Time (ms)")
plt.xticks([0, 1], ["Incongruent", "Congruent"])
plt.xlabel("")
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

A paired t-test revealed a significant difference in reaction times between congruent and incongruent conditions (t(n1) = 2.77, p = 0.007). This was further supported by a Wilcoxon signed-rank test, which also showed a significant difference (W = 547, p = 0.0026), suggesting the effect is robust even without assuming normality.

```python
congruent_rt = df[df["is_congruent"] == 1]["reaction_time"]
incongruent_rt = df[df["is_congruent"] == 0]["reaction_time"]
incongruent_rt
```

```python
from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(incongruent_rt, congruent_rt)
print(f"Independent T-test: t = {t_stat:.3f}, p = {p_value:.4f}") 
```

```python

sns.boxplot(x='is_congruent', y='reaction_time', data=df, palette="Set2")
plt.title("Reaction Time by Condition")
plt.ylabel("Reaction Time (ms)")
plt.xlabel("Is Congruent")
plt.show()
```

**Trends over time:**

```python
df["trial_index"] = df.groupby("participant_id").cumcount() + 1
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="trial_index", y="reaction_time", hue="participant_id", alpha=0.4, legend=False)
plt.title("Reaction Time Trends Per Participant")
plt.xlabel("Trial Number")
plt.ylabel("Reaction Time (ms)")
plt.show()
```

```python
# Add slope per participant using linear regression
from scipy.stats import linregress

slopes = []
for pid, group in df.groupby("participant_id"):
    slope, _, _, p_val, _ = linregress(group["trial_index"], group["reaction_time"])
    slopes.append({"participant_id": pid, "slope": slope, "p_value": p_val})

trend_df = pd.DataFrame(slopes)
print(trend_df.head())
```

```python
num_faster = (trend_df["slope"] < 0).sum()
num_slower = (trend_df["slope"] > 0).sum()

print(f"Participants getting faster: {num_faster}")
print(f"Participants getting slower: {num_slower}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.hist(trend_df["slope"], bins=15, color="skyblue", edgecolor="black")
plt.axvline(0, color="red", linestyle="--")
plt.title("Distribution of RT Trend Slopes Across Participants")
plt.xlabel("Slope (ms per trial)")
plt.ylabel("Number of Participants")
plt.show()
```

No strong overall trend: With 30 getting faster and 34 slower, theres no dominant direction  suggests individual differences in attention, fatigue, or learning during the Stroop test.

T-test result (if p > 0.05) would confirm this  no consistent trend across participants.

This adds nuance to your project  instead of saying people got faster/slower, you can say:

We observed a near-even split in slope direction across participants, with no statistically significant group-wide trend in reaction time change over trials. This highlights strong individual variability in how participants adapt (or not) over repeated Stroop trials.

**Heatmap:**

```python
# Create confusion matrix
conf_matrix = pd.crosstab(df["color"], df["response"])

# Optional: normalize by row to get proportions
conf_matrix_normalized = conf_matrix.div(conf_matrix.sum(axis=1), axis=0)
```

```python
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix_normalized, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title("Confusion Matrix: Font Color vs Response")
plt.xlabel("Responded As")
plt.ylabel("Actual Font Color")
plt.show()
```

Seems to have lowest accuracy with green(0.89).
highest with yellow(0.95)

"Among the four colors, green showed the lowest response accuracy (89%), while yellow had the highest (95%). This may reflect perceptual confusion or stronger interference effects for the word 'GREEN'. Most errors consisted of confusing blue with green, and green with yellow (both 5%)."
